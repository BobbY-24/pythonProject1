```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import yfinance as yf

# Load data
DATA_DIR = Path("../src/data")
df = pd.read_parquet(DATA_DIR / "AAPL.parquet")
df = df.reset_index()


# Quick look
print(df.head())
print(df.describe())

# Plot closing price
df.plot(x="Date", y="Close", title="AAPL Closing Price", figsize=(10, 5))
plt.show()

# Volume distribution
df["Volume"].plot(kind="hist", bins=50, title="AAPL Volume Distribution", figsize=(8, 5))
plt.show()
```
